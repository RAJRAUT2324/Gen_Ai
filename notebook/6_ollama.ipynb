```python
from langchain_ollama import ChatOllama
from pydantic import BaseModel, Field

# ==========================================
# APPROACH 1: Standard Chat / Text Output
# ==========================================
llm = ChatOllama(model="llama3", temperature=0)

prompt = [
    ("system", "You are a C++ developer."),
    ("user", "Which library file is added at the top for input/output?")
]

text_res = llm.invoke(prompt)
print("--- Standard Output ---")
print(text_res.content)


# ==========================================
# APPROACH 2: Structured Output (Pydantic)
# ==========================================
# Define your schema
class MovieInfo(BaseModel):
    movie: str = Field(description="The name of the movie")
    year: int = Field(description="The release year of the movie")

# Bind the Pydantic schema to ChatOllama
structured_llm = llm.with_structured_output(MovieInfo)

movie_prompt = [
    ("system", "You are a movie expert that extracts movie details accurately."),
    ("user", "Tell me about Interstellar.")
]

structured_res = structured_llm.invoke(movie_prompt)

print("\n--- Structured Output ---")
print(f"Movie: {structured_res.movie}")
print(f"Year: {structured_res.year}")
```
