```python
from dotenv import load_dotenv

load_dotenv(override=True) # override= true can be useed when jupiter book not restart 
import os
print(os.getenv("SERPER_API_KEY"))
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain_groq import ChatGroq
from langchain.agents import create_agent
```

```python

llm=ChatGroq(model="openai/gpt-oss-20b")
search=GoogleSerperAPIWrapper()
```

```python
agent=create_agent(
    model=llm,
    tools=[search.run],
    system_prompt="You are aagent who can search any question on google"
)

question="who is president of united state"
response=agent.invoke({"messages":[{"role":"user","content":question}]})
print(response["messages"][-1].content)
```
