```python
from dotenv import load_dotenv
load_dotenv()
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
prompt=[
    { "role":"system","content":"you are cricket coach"},
    { "role":"user","content":"give me rules of how cricket works"}
]
res = llm.invoke(prompt)
print(res.content)
```

```python
#Templete prompt chata and dynamic prompts
from dotenv import load_dotenv
load_dotenv()
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a {language} coach."),
    ("user", "{query}")
])
prompt_p=prompt.format_messages(language="kabaddi",query="how many people")
res=llm.invoke(prompt_p)
print(res.content)
```
