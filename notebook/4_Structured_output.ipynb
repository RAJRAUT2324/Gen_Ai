```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel, Field

# Load environment variables
load_dotenv()

# 1. Define your desired output structure using Pydantic
class MovieInfo(BaseModel):
    movie: str = Field(description="The name of the movie")
    year: int = Field(description="The release year of the movie")

# 2. Initialize the model
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# 3. Instruct the LLM to structure its response according to your schema
structured_llm = llm.with_structured_output(MovieInfo)

# 4. Define your prompt
prompt = [
    ("system", "You are a movie expert that extracts or provides movie details accurately."),
    ("user", "Tell me about the movie Inception.")
]

# 5. Invoke the structured model
res = structured_llm.invoke(prompt)

# 'res' is now an instance of the MovieInfo Pydantic class
print(f"Movie: {res.movie}")
print(f"Year: {res.year}")
print(f"JSON Output: {res.model_dump_json(indent=2)}")
```
